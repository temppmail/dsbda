```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./Iris.csv',header=None)
df.head()
```

```python
df.columns = ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm','Species']
df.head()
```

## 1. Create data subsets for different species


```python
df['Species'].unique()
```

```python
df_setosa = df[df['Species'] == 'Iris-setosa']
df_setosa.head()
```

```python
df_versicolor = df[df['Species'] == 'Iris-versicolor']
df_versicolor.head()
```

## 2. Merge two subsets

```python
df_merge = pd.concat([df_setosa,df_versicolor])
df_merge.head()
```

## 3. Sort Data Petal Length


```python
df_sort = df.sort_values('PetalLengthCm',ascending=False)
df_sort.head()
```

## 4. Transposing Data


```python
df_transpose = df.T
df_transpose.head()
```

## 5. Melting Data to long format


```python
df_melt = pd.melt(df,id_vars=['SepalLengthCm','PetalLengthCm','Species'])
df_melt.head()
```

## 6. Casting data to wide format

```python
df_cast = pd.pivot_table(df,index=['Species'])
df_cast.head()
```
