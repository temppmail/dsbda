```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
#reading the datasets
# Switzerland hung
df = pd.read_csv("Cleavland.csv",header=None)
df.head()
```

```python
df.columns = ['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal','num']
df.head()
```

## Data Cleaning

```python
df.isnull().sum().sum()
```

```python
(df=='?').sum().sum()
```

```python
df.shape
```

```python
df = df.replace('?', np.nan)
# df = df.dropna()
df = df.fillna(df.mean())
df.head()
```

```python
df.shape
```

# Plotting the graphs


## 1. Histogram

### Objective: Visualize the distribution of ages in the dataset

```python
plt.figure(figsize=(8, 6))
plt.hist(df['age'], bins=10, color='skyblue')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution') 
plt.show()
```

## 2. Pie Charts

### Objective: Compare the distribution of sexes in the dataset

```python
plt.figure(figsize=(8, 6))
df['sex'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightpink'])
plt.ylabel('')
plt.title('Sex Distribution')  # Objective: Compare the distribution of sexes in the dataset
plt.show()
```

## 3. Box Plots
### Objective: Compare cholesterol levels for different chest pain types

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='cp', y='chol', data=df)
plt.xlabel('Chest Pain Type')
plt.ylabel('Cholesterol')
plt.title('Chest Pain vs. Cholesterol')  
plt.show()
```

## 4. Scatter Plots
### Objective: Explore the relationship between age and heart rate

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='age', y='thalach', data=df, hue='num')
plt.xlabel('Age')
plt.ylabel('Heart Rate')
plt.title('Age vs. Heart Rate')  
plt.show()
```

## 5. Bar Plot
### Objective: Compare the frequency of different types of heart disease

```python
plt.figure(figsize=(8, 6))
df['num'].value_counts().sort_index().plot(kind='bar', color='lightgreen')
plt.xlabel('Heart Disease')
plt.ylabel('Frequency')
plt.title('Heart Disease Distribution') 
plt.show()
```
